```python
import pandas as pd 
from matplotlib import pylot as plt
import seaborn as sns 
```

```python
titanic = pd.read_csv('titanic_clean.csv')
```

```python
titanic.head()
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare'
)
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3
)
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3,
    hue = 'sex'
    
)
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3,
    hue = 'sex',
    col = 'survived'
)
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3,
    hue = 'sex',
    col = 'survived',
    palette={"male":"yellow", "female":"magenta"}
)
```

```python
titanic['Sex'] = titanic['sex'].map({"male":"Male","female":"Female"})

sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3,
    hue = 'Sex',
    col = 'survived',
    palette={"Male":"yellow", "Female":"magenta"}
)
```

```python
sns.relplot(
    data=titanic, 
    x = "age", y = 'fare',
    alpha = .3,
    hue = 'Sex',
    col = 'survived',
    palette={"Male":"yellow", "Female":"magenta"},
    size="pclass"
)
```

```python
mental.groupby("Choose your gender")[['Do you have Depression?','Do you have Anxiety?','Do you have Panic attack?']].count()
```

```python
columns_to_plot = mental.columns[-4:-1]

category_order = ['Yes', 'No']
```

```python
for column in columns_to_plot:
    counts = mental[column].value_counts()

    counts = counts.reindex(category_order, fill_value=0)

    print(counts)
    print()
```
